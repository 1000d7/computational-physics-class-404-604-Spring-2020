```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def derifw(y,x,dx):
    """
    forward difference formula
    
    """
    return (y(x+dx)-y(x))/dx

def dericd(y,x,dx):
    """
    central difference formula
    """
    return (y(x+dx/2)-y(x-dx/2))/dx

```

$$\frac{d}{dx}\cos{x} = -\sin{x}$$

$$-\sin{\frac{\pi}{4}} = -\frac{1}{\sqrt{2}}$$

$$-\frac{1}{\sqrt{2}} \approx -0.7071067811865475 $$

# 1)

```python
yarr = [] # array for y values
dxarr = [] # array for differential values
residuals = [] # array for residuals ( abs( a0 - y ) ) 
a0 = -1/np.sqrt(2) # analytic solution for the derivative of cos(pi/4)
for i in range(1, 16): 
    """
    Calculate the numeric derivative of cos(pi/4) then find residuals
    and append values to arrays
    
    """
    dx = 10**-i
    y = derifw(np.cos, np.pi/4, dx)
    r = np.abs(y - a0)
    residuals.append(r)
    dxarr.append(dx)
    yarr.append(y)
    iarr.append(i)

# aggrigate data into dataframe
df = {'dx':dxarr, 'y':yarr, 'r':residuals}      
df = pd.DataFrame(df)
print(df)

# plot of residuals vs dx
plt.plot(dxarr, residuals)
plt.xscale('log')
plt.yscale('log')
plt.title('log residual plot for the foward difference')
plt.ylabel('$y_{analytic} - y_{numeric}$')
plt.xlabel('dx')
```

### Using the foward difference algorithem we find an optimal differential size of 10^-8 that minimize the error
### to approximately 10^-8
